# visualizing data using Pandas

## Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('emp.csv')
```

```python
df.head()
```

```python
df.plot(kind='scatter',x='exp',y='salary', color='blue')
plt.title("Experience Vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary in Dollars")
plt.show()
```

```python
plt.scatter(df['exp'],df['salary'], color='green', marker='x',s=100,alpha=0.7)
plt.title("Experience Vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary in Dollars")
plt.show()
```

```python
colors=df['emp_type'].map({'Private':'red','Government':'blue'})

plt.scatter(df['exp'],df['salary'], c=colors, marker='x',s=100,alpha=0.7)
plt.title("Experience Vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary in Dollars")
plt.show()
```

```python
import matplotlib.patches as mpatches

plt.scatter(df['exp'],df['salary'], c=colors, marker='x',s=100,alpha=0.7)
plt.title("Experience Vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary in Dollars")

red_patch=mpatches.Patch(color='red', label='Private')
blue_patch=mpatches.Patch(color='blue', label='Government')
plt.legend(handles=[red_patch,blue_patch])

plt.show()
```
