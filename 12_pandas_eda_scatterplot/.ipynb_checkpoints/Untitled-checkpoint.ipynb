```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Experience': [1, 2, 3, 4, 5, 6, 7, 8],
    'Salary': [30, 35, 40, 45, 50, 55, 60, 65]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.plot(kind='scatter', x='Experience', y='Salary', color='blue')
plt.title("Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (in thousands)")
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Experience'], df['Salary'], 
            color='green', 
            marker='o', 
            s=100, 
            alpha=0.9)

plt.title("Customized Scatter Plot")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
df['Department'] = ['IT', 'HR', 'IT', 'HR', 'IT', 'HR', 'IT', 'HR']
colors = df['Department'].map({'IT': 'red', 'HR': 'blue'})

plt.scatter(df['Experience'], df['Salary'], c=colors)
plt.title("Scatter Plot by Department")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'age': [19, 18, 28, 33, 32],
    'sex': ['female', 'male', 'male', 'male', 'male'],
    'bmi': [27.9, 33.77, 33.0, 22.705, 28.88],
    'children': [0, 1, 3, 0, 0],
    'smoker': ['yes', 'no', 'no', 'no', 'no'],
    'region': ['southwest', 'southeast', 'southeast', 'northwest', 'northwest'],
    'charges': [16884.924, 1725.5523, 4449.462, 21984.47061, 3866.8552]
}

df4 = pd.DataFrame(data)
print(df4)
```

```python
plt.scatter(df4['age'], df4['charges'], 
            color='green', 
            marker='o', 
            s=100, 
            alpha=0.6)

plt.title("Age VS Insurance Charges")
plt.xlabel("Age")
plt.ylabel("charges")
plt.grid(True)
plt.show()
```
