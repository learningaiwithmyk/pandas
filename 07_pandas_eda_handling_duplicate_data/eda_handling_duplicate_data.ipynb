```python
import pandas as pd
df1=pd.read_json('house_data.json')
```

```python
#df.drop_duplicates()
df2=df1.drop_duplicates()
```

```python
df2
```

```python
df2.reset_index(drop=True)
df2
```

```python
df2.reset_index(drop=True,inplace=True)
```

```python
df2
```

```python
#df.replace(old_value,new_value)
df2['car_parking']=df2['car_parking'].replace('Not Available','Available')
```

```python
df2
```

```python
df2['location']=df2['location'].replace({'Boston':'Bt','Portland':'PL'})
df2
```

```python
df2['location']=df2['location'].replace(r'San.*','SF', regex=True)
```
