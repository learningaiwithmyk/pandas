## Handling Missing Data in Pandas

#### #1 How to Identify Null/NA Values in DF

#### #2 How to remove Null/NA Values in DF

#### #3 How to Replace Missing Values in DF

```python
#Data file financial_data.xlsx
```

```python
import pandas as pd
```

```python
df=pd.read_excel('financial_data.xlsx')
```

```python
df.shape
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#df.dropna()
df2=df.dropna()
```

```python
df2
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
df['Discounts']=df['Discounts'].fillna(df['Discounts'].mean())
```

```python
df.isnull().sum()
```

```python
df['Profit']=df['Profit'].fillna(df['Profit'].median())
```

```python
df.isnull().sum()
```

```python
df['Discount Band']=df['Discount Band'].fillna('unknown')
```
