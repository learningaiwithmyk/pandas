## Filtering & Sorting Data in Pandas

### #1 Filtering Data in Pandas

```python
import pandas as pd
df=pd.read_csv('students.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
age_df=df[df['Age']>22]
age_df
```

```python
age_gp_df=df[(df['Age']>22)&(df['GPA']>3)]
age_gp_df
```

```python
age_gp_df=df[(df['Age']>22)|(df['GPA']>3)]
age_gp_df
```

```python
gy_df=df[~(df['GraduationYear']==2025)]
gy_df
```

```python
gy_df=df[df['GraduationYear']!=2027]
gy_df
```

### #2 Sorting Data in Pandas

```python
age_df=df.sort_values(by='Age')
age_df
```

```python
age_df.sort_values(by='Age', ascending=False)
```

```python
age_df.sort_values(by=['Age','GraduationYear'], ascending=[False,False])
```

```python
age_df.sort_values(by=['Age','GraduationYear'], ascending=[False,True])
```

```python
age_df.sort_values(by='Age', na_position='first')
```

```python
age_df.sort_values(by='GraduationYear', kind='mergesort')
```

```python
email_df=df.sort_values(by='Email')
email_df
```

```python
email_df=df.sort_values(by='Email' , key=lambda col: col.str.split("@").str[1])
email_df
```
